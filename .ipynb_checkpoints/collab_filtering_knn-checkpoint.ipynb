```python
# Imports

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import sklearn
import catboost
```

```python
# Loading Data

train = pd.read_csv('data/train.csv')
songs = pd.read_csv('data/songs.csv')
song_labels = pd.read_csv('data/song_labels.csv')
test = pd.read_csv('data/test.csv')
save_for_later = pd.read_csv('data/save_for_later.csv')
dummy_submission = pd.read_csv('data/dummy_submission.csv')
```

```python
matrix_nan = train.pivot_table(index = 'customer_id', columns = 'song_id', values = 'score')
```

```python
matrix_mean = matrix_nan - matrix_nan.mean()
```

```python
matrix_mean.fillna(0, inplace = True)
```

```python
corr_mat = np.corrcoef(matrix_mean.to_numpy(), rowvar = False)
```

```python
matrix_mean
```

```python
corr_mat.shape
```

```python
train['song_id'].unique()
```

```python
train['song_1'] = np.zeros(len(train.index.to_numpy()))
train['song_2'] = np.zeros(len(train.index.to_numpy()))
train['song_3'] = np.zeros(len(train.index.to_numpy()))

train['song_1_coeff'] = np.zeros(len(train.index.to_numpy()))
train['song_2_coeff'] = np.zeros(len(train.index.to_numpy()))
train['song_3_coeff'] = np.zeros(len(train.index.to_numpy()))
```

```python
train
```

```python
customer_groups = train.groupby('customer_id')
```

```python
for i in range(len(train.index.to_numpy())):
    customer = train['customer_id'].iloc[i]
    song = train['song_id'].iloc[i]
    group = customer_groups.get_group(customer)
    customer_songs = group['song_id'].unique()
    customer_song_loc = customer_songs-1
    coeffs = corr_mat[customer_song_loc][:, song-1]
    indices = (-coeffs).argsort()[1:4]
    max_coeffs = coeffs[indices]
    nearest_songs = customer_songs[indices]
    train['song_1'].iloc[i] = nearest_songs[0]
    train['song_2'].iloc[i] = nearest_songs[1]
    train['song_3'].iloc[i] = nearest_songs[2]
    train['song_1_coeff'].iloc[i] = max_coeffs[0]
    train['song_2_coeff'].iloc[i] = max_coeffs[1]
    train['song_2_coeff'].iloc[i] = max_coeffs[2]
    if i%100==0:
        print(i)
```
